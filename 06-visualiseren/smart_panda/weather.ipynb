## Minmax and Pandas

Wat is de maximale (hoogst) gemiddelde temperatuur in Rotterdam?

Stap 1, we halen Rotterdam uit tabel dataste en zetten deze in een nieuwe dataset.

```python
temp_rot = weather_nl[(weather_nl["City"] == "Rotterdam") & (weather_nl["AverageTemperature"])]
temp_rot
```

Stap 2, We halen de maximale temperatuur op in Rotterdam.

```python
column = temp_rot["AverageTemperature"]
temp_rot_max = column.max()
print("Maximale temperatuur:",temp_rot_max)
```

Stap 1, we halen Maastrciht uit de dataset en zetten deze in een nieuwe dataset.

```python
temp_maas = weather_nl[(weather_nl["City"] == "Maastricht") & (weather_nl["AverageTemperature"])]
temp_maas
```

Stap 2, we halen de minimale temperatuur op in Maasstricht.

```python
column = temp_maas["AverageTemperature"]
temp_maas_min = column.min()
print("Minimale temperatuur:",temp_maas_min)
```

Hoeveel verschillende steden zitten in de dataset?

```python
print("Aantal steden: ", weather_nl['City'].nunique())
```

Welke gemiddelde temperatuur komt het meeste voor? Rond hiervoor de temperatuur af op hele getallen.

```python
temp_meest = weather_nl["AverageTemperature"].mode().max().round(decimals=0).astype(int)
print("Meest voorkomende temperatuur:", temp_meest, "graden")
```

Wat zijn de top 3 warmste temperaturen gemeten? Sommige temperaturen kunnen meer dan één keer voorkomen.

```python
temp_warmst = weather_nl.nlargest(4, "AverageTemperature").drop_duplicates(subset=["AverageTemperature"])
val_temp = temp_warmst["AverageTemperature"].values[0:3]
val_temp
print("3 warmste temperaturen:", val_temp)
```
